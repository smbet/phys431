# Now doing thermal noise or whatever:

```python
thermal_noise = [
    [
        [9.99],[4.127,4.187,4.160,4.050,4.132],[2.242,2.257,2.222,2.205,2.256]
    ],
    [
        [7.50],[3.576,3.594,3.626,3.605,3.649],[1.977,1.927,1.943,1.967,1.983]
    ],
    [
        [4.99],[2.967,3.019,2.988,3.011,2.985],[1.659,1.652,1.678,1.652,1.659]
    ],
    [
        [2.50],[2.151,2.178,2.173,2.220,2.164],[1.264,1.243,1.257,1.251,1.236]
    ],
    [
        [1.00],[1.447,1.428,1.446,1.438,1.435],[0.934,0.9188,0.9385,0.9318,0.9250]
    ],
    [
        [0],[0.6277,0.6235,0.6305,0.6323,0.6255],[0.6091,0.6058,0.6034,0.6000,0.5836]
    ],
]

# okay so this one goes: resistance, room temp, 77 K

thermal_noise = np.array(thermal_noise).T
```

```python
# not the cleanest code I've ever written, but I just needed something working

resistance = np.array([thermal_noise[0][ii] for ii in range(len(thermal_noise[0]))]).reshape(len(thermal_noise[0]),) # k ohms
noise_room_temp = np.array([np.median(thermal_noise[1][ii]) for ii in range(len(thermal_noise[1]))]).reshape(len(thermal_noise[1]),)**2
noise_77k = np.array([np.median(thermal_noise[2][ii]) for ii in range(len(thermal_noise[2]))]).reshape(len(thermal_noise[2]),)**2
```

```python
# finding linear fit for data:

b1,m1 = polyfit(resistance, noise_room_temp, deg=1)
b2,m2 = polyfit(resistance, noise_77k, deg=1)
```

```python
plt.errorbar(resistance, noise_room_temp, xerr=0.01, yerr=0.1,fmt='o', capsize=3.5, c='xkcd:Barbie Pink', alpha=0.8, label='Median Room Temp')
plt.errorbar(resistance, noise_77k, xerr=0.01, yerr=0.1,fmt='o', capsize=3.5, c='xkcd:Cool Blue', alpha=0.8, label='Median 77 K')
plt.plot(resistance, m1*resistance+b1, color='xkcd:Barbie Pink', alpha=0.6, label='Linear Median Fit Room Temp') # Come on Barbie, let's go party!
plt.plot(resistance, m2*resistance+b2, color='xkcd:Cool Blue', alpha=0.6, label='Linear Fit 77 K')
plt.legend(loc=2)
plt.xlabel('Resistance (k $\Omega$)')
plt.ylabel('Potential (mV)')
plt.title("Median Thermal Noise")
plt.savefig('./noize_effect_thermal_median');
```
