```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt
import astropy.units
import astropy.constants as co

from numpy.polynomial.polynomial import polyfit
```

```python
plt.style.use('default')
plt.style.use('ggplot')
```

```python
gollllddddddd = [[40.013, 
-0.03061, 
-0.03120], 
[50.041, 
-0.03888, 
-0.03960],
[60.100, 
-0.04646, 
-0.04772], 
[66.055, 
-0.05201, 
-0.05290], 
[70.316, 
-0.05586, 
-0.05687], 
[125.11, 
-0.1015, 
-0.10325]]

gollllddddddd = np.array(gollllddddddd).T
```

```python
gollllddddddd.shape
```

```python
current = gollllddddddd[0]
hall = (gollllddddddd[1]-gollllddddddd[2])*10**(3) # converting from mV to uV
```

```python
# finding linear fit for data:

b,m = polyfit(current, hall, deg=1)
```

```python
hall
```

```python
plt.scatter(current, hall)
plt.errorbar(current, hall, yerr=2.0,fmt='o', capsize=5, c='xkcd:black', alpha=0.6, label='Measurements')
plt.plot(current, m*current+b, color='xkcd:Barbie Pink', alpha=0.6, label='Linear Fit') # Come on Barbie, let's go party!
plt.ylim(0,5)
plt.legend(loc=4)
plt.xlabel('Current (mA)')
plt.ylabel('Potential ($\mu$V)')
plt.title("There's a Hall Effect (Au Sample)")
plt.savefig('./Hall_effect_gold');
```

```python
print(m, b)
```

Line is potential ($\mu$V) = 0.013 * current (mA) + 0.15 ish
