## Finally calculating the actual Johnson noise:

```python
def johnsonerator(R,V2,T,c):
    kayy=(V2-c)/(10000**2*4*10*R*T)
    return(kayy)
```

```python
def justJohnson(Vobs,c,gain):
    V = (Vobs-c)/(gain**2)
    return(V)
```

```python
plt.plot(resistance, justJohnson(b1*resistance+c1,0.39400729,10000), color='xkcd:Barbie Pink', alpha=0.6, label='Johnson Fit Room Temp') # Come on Barbie, let's go party!
plt.plot(resistance, justJohnson(b2*resistance+c2,c2,10000), color='xkcd:Cool Blue', alpha=0.6, label='Johnson Fit 77 K')
plt.errorbar(janky_resistance_array, justJohnson(janky_room_temp,0.39400729,10000), xerr=0.01, yerr=0.01*10**(-7),fmt='o', capsize=3.5, c='xkcd:Barbie Pink', alpha=0.6, label='Room Temp')
plt.errorbar(janky_resistance_array, justJohnson(janky_77k,c2,10000), xerr=0.01, yerr=0.01*10**(-7),fmt='o', capsize=3.5, c='xkcd:Cool Blue', alpha=0.6, label='77 K')
plt.legend(loc=2)
plt.xlabel('Resistance (k$\Omega$)')
plt.ylabel('Square of Potential (mV$^2$)')
plt.title("Square of Johnson Noise")
plt.savefig('./johnson_noizzzze');
```
